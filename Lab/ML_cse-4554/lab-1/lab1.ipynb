```python
import pandas as pd

# Load the California housing dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot a heatmap for visualizing correlation
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['longitude'], df['latitude'], c=df['median_house_value'], cmap='viridis', marker='o', alpha=0.5)
plt.colorbar(label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of House Prices')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, c='blue')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Median Income vs House Value')
plt.show()
```

```python
# Sort by median house value
top_locations = df[['longitude', 'latitude', 'median_house_value', 'median_income']].sort_values(by='median_house_value', ascending=False).head(10)
print(top_locations)
```
